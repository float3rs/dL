```python
import pandas
import calendar
import matplotlib
from matplotlib import pyplot
```

```python
%matplotlib inline
```

```python
df = pandas.read_csv('full.txt', header=None, skiprows=2, sep='\s+', names=['year', 'month', 'day', 'hour', 'minute', 'second', 'latitude', 'lotitude', 'depth', 'magnitude'])
```

```python
df
```

```python
df.index.name = 'id'
```

```python
df
```

```python
df['month'] = df['month'].str.title()
```

```python
df
```

```python
month_abbr_as_ints = dict((x,y) for (y,x) in enumerate(calendar.month_abbr))
month_abbr_as_ints
```

```python
df['month'].replace(month_abbr_as_ints, inplace=True)
```

```python
df
```

```python
df['datetime'] = (
    df['year'].astype(str) + '-' + 
    df['month'].astype(str) + '-' +
    df['day'].astype(str) + ' ' +
    df['hour'].astype(str) + ':' +
    df['minute'].astype(str) + ':' +
    df['second'].astype(str)
)
    
```

```python
df
```

```python
df.drop(columns = ['year', 'month', 'day', 'hour', 'minute', 'second'], inplace = True)
```

```python
df
```

```python
list(df.columns.values)
```

```python
df = df[['datetime', 'latitude', 'lotitude', 'depth', 'magnitude']]
```

```python
df
```

```python
df.dtypes
```

```python
df['datetime'] = pandas.to_datetime(df['datetime'])
```

```python
df
```

```python
df.dtypes
```

```python
datetimes = df['datetime']
```

```python
datetimes
```

```python
events = list(df.index)
```

```python
pyplot.plot(datetimes, events)
pyplot.title('cumulative events by datetime')
pyplot.xlabel('datetime')
pyplot.ylabel('cumulative events')
```

```python
magnitudes = df['magnitude']
```

```python
pyplot.plot(datetimes, magnitudes)
pyplot.title('magnitude by datetime')
pyplot.xlabel('datetime')
pyplot.ylabel('magnitude')
```

```python
pyplot.plot(datetimes, events)
pyplot.title('cumulative events by datetime')
pyplot.xlabel('datetime')
pyplot.ylabel('cumulative events')
pyplot.plot(datetimes, magnitudes)
pyplot.title('magnitude by datetime')
pyplot.xlabel('datetime')
pyplot.ylabel('magnitude')
pyplot.legend(['cumulative events', 'magnitudes'])
```

```python
pyplot.plot(datetimes, events, label='cumulative events')
pyplot.title('cumulative events by datetime')
pyplot.xlabel('datetime')
pyplot.ylabel('cumulative events')
pyplot.plot(datetimes, magnitudes, label='magnitudes')
pyplot.title('magnitude by datetime')
pyplot.xlabel('datetime')
pyplot.ylabel('magnitude')
pyplot.legend()
```

```python
pyplot.plot(df_1990s.index, df_1990s['event'])
pyplot.title('cumulative events by datetime')
pyplot.xlabel('datetime')
pyplot.ylabel('cumulative events')
```

```python
pyplot.plot(df_2000s.index, df_2000s['event'])
pyplot.title('cumulative events by datetime')
pyplot.xlabel('datetime')
pyplot.ylabel('cumulative events')
```

```python
pyplot.plot(df_1990s.index, df_1990s['event'], label='1990s')
pyplot.title('cumulative events by datetime')
pyplot.xlabel('datetime')
pyplot.ylabel('cumulative events')
pyplot.plot(df_2000s.index, df_2000s['event'], label='2000s')
pyplot.title('cumulative events by datetime')
pyplot.xlabel('datetime')
pyplot.ylabel('cumulative events')
pyplot.legend(loc='upper left')
# HOW TO SAVE THE FIGURE PROGRAMMATICALY
pyplot.savefig('plot.png')
```

```python
df_2000s['event'].resample('Y').plot()
pyplot.title('cumulative events by datetime')
pyplot.xlabel('datetime')
pyplot.ylabel('cumulative events')
```

```python
df_1990s['event'].resample('Y').plot(label='1990s')
pyplot.title('cumulative events by datetime')
pyplot.xlabel('datetime')
pyplot.ylabel('cumulative events')
df_2000s['event'].resample('Y').plot(label='2000s')
pyplot.title('cumulative events by datetime')
pyplot.xlabel('datetime')
pyplot.ylabel('cumulative events')
pyplot.tight_layout()
pyplot.grid(True)
pyplot.legend()
```

```python
df_orpheus = df['2021-01-18 00:00:00':'2021-01-18 23:59:59'].copy()
```

```python
df_orpheus.count()
```

```python
pyplot.rcdefaults()
```

```python
pyplot.style.use('seaborn')
pyplot.plot(df_orpheus.index, df_orpheus['magnitude'])
pyplot.title('earthquake magnitudes the day Orpheus was born')
pyplot.xlabel('datetime')
pyplot.ylabel('magnitude')
pyplot.grid(True)
```

```python
pyplot.style.use('seaborn')
pyplot.scatter(df_orpheus.index, df_orpheus['magnitude'], s=16, marker='o', 
    edgecolors='black', linewidth=0.1, alpha=0.75) # s: size
pyplot.title('earthquake magnitudes the day Orpheus was born')
pyplot.xlabel('datetime')
pyplot.ylabel('magnitude')
pyplot.grid(True)
```

let's also assign the depth

```python
df_orpheus['depth']
```

```python
colors = list(df_orpheus['depth'])
```

```python
pyplot.style.use('seaborn')
pyplot.scatter(df_orpheus.index, df_orpheus['magnitude'], s=16, marker='o', 
    edgecolors='black', linewidth=0.1, alpha=0.90, c=colors) # s: size, c: colors
pyplot.title('earthquake magnitudes the day Orpheus was born')
pyplot.xlabel('datetime')
pyplot.ylabel('magnitude')
pyplot.grid(True)
```

we can use a color map

```python
pyplot.style.use('seaborn')
pyplot.scatter(df_orpheus.index, df_orpheus['magnitude'], s=16, marker='o', 
    edgecolors='black', linewidth=0.1, alpha=0.90, cmap='Greens', c=colors) # s: size, c: colors
pyplot.title('earthquake magnitudes the day Orpheus was born')
pyplot.xlabel('datetime')
pyplot.ylabel('magnitude')
pyplot.grid(True)
cbar = pyplot.colorbar()
cbar.set_label('depth (km)')
```

```python
df_orpheus['depth']
```

```python
sizes = list(df_orpheus['magnitude'])
colors = list(df_orpheus['latitude'])
```

```python
pyplot.style.use('seaborn')
sctr = pyplot.scatter(df_orpheus.index, df_orpheus['depth'], s=sizes, marker='o', 
    edgecolors='black', linewidth=0.1, alpha=0.90, c=colors) # s: size, c: colors
pyplot.title('earthquakes the day Orpheus was born')
pyplot.xlabel('datetime')
pyplot.ylabel('depth')

sctr.axes.invert_yaxis()

# could also be done with the gcf() method (get currebt figure)
# pyplot.gcf().axes.invert_axis()

pyplot.colorbar().set_label('Latitude')

pyplot.grid(True)
```

log Scale!

```python
pyplot.style.use('seaborn')
pyplot.scatter(df_orpheus.index, df_orpheus['magnitude'], s=sizes, marker='o', 
    edgecolors='black', linewidth=0.1, alpha=0.90) # s: size, c: colors

pyplot.yscale('log')

pyplot.title('earthquake magnitudes the day Orpheus was born')
pyplot.xlabel('datetime')
pyplot.ylabel('magnitude')
pyplot.grid(True)
```
